````python
'''
Given a topic, this gives a suggestion of N (number of results) linked words
related to the topic
'''
def suggestion(topic,N):
  results=wikipedia.search(topic, N)
  return results
'''
We are going to fill the empty list in each value of the dictionary with the
suggestion words related to the key word.
'''
def fill_list_dict(dictionary,N):
  keys=dictionary.keys()
  for key in keys:
    list_suggestions=suggestion(key, N)
    dictionary[key]=list_suggestions
    #print(list_by_key)
  #print(keys)
  return dictionary
#list_dict(dictionario_busqueda)
````

````python
#Dictionary to fill with related word with the previous methods
dictionary_search={
      'Macroeconomía' : [],
      'Sostenibilidad' : [],
      'Innovación': [],
      'Regulaciones' : [],
      'Alianzas' : [],
      'Reputación' : []}


fill_list_dict(dictionary_search,5)
````

According to the previous list in each value of the dictionary, we want to obtain all the information with the suggested words and create a dictionary with the wikipedia content of each topic.

````python
'''
---- Optional: ----
To create a dictionary with values empty lists
'''
def dic_empty_lists(dic_wor_sug):
  new_dict={key : []  for key in dic_wor_sug}
  return new_dict

'''
We are going to create a new dictionary with information given by the suggested
words given in the dictionary with the words related to the key
'''
def fill_dict_sug_words(dic_wor_sug):
  new_dict={key : []  for key in dic_wor_sug}
  #vanish the comment if you want to change the searching language and change the code
  #wikipedia.set_lang('es')
  for key in dic_wor_sug:
    for word in dic_wor_sug[key]:
      try:
        page=wikipedia.page(word)
        content=page.content
        new_dict[key].append(content)
      except:
        continue
  return new_dict

new_dict=fill_dict_sug_words(dictionary_search)
new_dict
````
