## **Use of .iloc and .loc**


Let us use another library called `seaborn` an let us import a dataset in order to study pandas

```python
from seaborn import load_dataset
df = load_dataset("exercise")
#This command print the first five rows
df.head()
```

### ***.iloc***
The method .iloc allows us to choose rows or columns according to its location. When we use .iloc on a dataframe, we obtain a Pandas.Series:

#### **To choose some rows:**

```python
#To obtain the first two rows
df.iloc[0:2]
```

```python
#To obtain the rows 2 and 3
df.iloc[2:4]
```

```python
#To obtain the rows 4, 7, 16 and 21
df.iloc[[4,7,16,21]]
```

```python
#To obtain the rows 4, 21, 7 and 16 (Note that the order is important)
df.iloc[[4,21,7,16]]
```

#### **To choose some columns:**

```python
#To obtain the first two columns
df.iloc[: , 0:2]
```

```python
#To obtain the columns 2 to 4
df.iloc[: , 2:5]
```

```python
#To obtain the columns 0, 3 and 5
df.iloc[: , [0,3,5]]
```

```python
#To obtain the columns 0, 5 and 3 (Note that the order is important)
df.iloc[: , [0,5,3]]
```

### ***.loc***
The methor .loc has two differents description:


1.   We can choose rows and columns based on a label.
2.   We can choose rows and columns based on a condition.

When we use .loc on a dataframe, we obtain a Pandas.Series:

```python
new_df=df.loc[4:10]
new_df
```

The following two codes are related to the use of .loc and .iloc in a dataframe filtered by means .loc method.

```python
'''
We are going to call the information according to the index
in the dataframe. located in the index 6
'''
new_df.loc[6]
```

```python
'''
We are going to call the information according to the position
in the dataframe. located in the position 6
'''
new_df.iloc[6]
```

#### ***.loc according to a label we can obtain a Pandas.Series***

```python
df.loc[: , 'id']
```

```python
df.loc[: , ['id', 'time', 'kind']]
```

```python
'''
The information give the rows from 20 to 25
and columns from 'id' to 'pulse'
'''
df.loc[20:25 , 'id':'pulse']
```

#### ***.loc according to a condition we can obtain a Pandas.Series***

```python
in_rest= df.loc[: , 'kind'] == 'rest'
in_rest
```

```python
df_in_rest=df[in_rest]
df_in_rest.head()
```
